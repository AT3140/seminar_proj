```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score, make_scorer
from sklearn.model_selection import train_test_split
```

```python
df1 = pd.read_excel('data.xlsx', sheet_name=0)
df2 = pd.read_excel('data.xlsx', sheet_name=1) 
```

```python
print(df1.columns)
print(df2.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1.loc[:,['Extrusion Temperature (in C)', 'Layer Height (mm)', 'Shell Thickness (mm)']], \
df1.loc[:, 'Expt'], train_size = 19, shuffle = False)
```

```python
from sklearn.cross_decomposition import PLSCanonical
pls2 = PLSCanonical(n_components=1)
pls2.fit(X_train, y_train)
# y_pred = pls2.predict(X_test)
y_pred = pls2.predict(df1.loc[:, ['Extrusion Temperature (in C)', 'Layer Height (mm)', 'Shell Thickness (mm)']])
```

```python
y_pred
```

```python
# result = df1.loc[df1['Dataset Type'] == 'Test', ['Dataset Type', 'Expt', 'XGBR']].reset_index(drop=True)
# result = df1.loc[df1['Dataset Type'] == 'Train', ['Dataset Type', 'Expt', 'XGBR']].reset_index(drop=True)
result = df1.loc[:, ['Dataset Type', 'Expt', 'XGBR']].reset_index(drop=True)
result['Predictions'] = pd.Series(y_pred)
result.to_csv('PLSCanonical.csv')
result
```

```python
print("r2 score (author) : ", r2_score(result['Expt'].to_numpy(), result['XGBR'].to_numpy()))
print("r2 score (amt) : ", r2_score(result['Expt'].to_numpy(), result['Predictions'].to_numpy()))
print(mean_squared_error(result['Expt'].to_numpy(), result['XGBR'].to_numpy()))
print(mean_squared_error(result['Expt'].to_numpy(), result['Predictions'].to_numpy()))
#print(pow(mean_squared_error(y_test.to_numpy(), predictions), 0.5))
```

```python
# Rewriting r2score definiton
def getRSquared(y, predictions):
    ssr = 0
    for i in range(len(y)):
        ssr = ssr + pow(y[i] - predictions[i], 2)
    sst = 0
    ymean = 0
    for y_i in y:
        ymean = ymean + y_i
    ymean = ymean / len(y)
    for i in range(len(y)):
        sst = sst + pow(y[i] - ymean, 2)
    return 1 - ssr / sst
```
